```python
notebook_name = "gpt_dev.ipynb"
target_language = 'ja'
```

```python
import json
from googletrans import Translator

def translate_to_korean(text, target_language=target_language):
    translator = Translator()
    try:
        translation = translator.translate(text, src='en', dest=target_language)
        return translation.text if translation is not None else None
    except TypeError:
        print(f"An error occurred while translating the text: {text}")
        return None

# Read the .ipynb file
with open(notebook_name, 'r', encoding='utf-8') as file:
    notebook = json.load(file)

# Iterate through the cells and translate markdown cells
for cell in notebook['cells']:
    if cell['cell_type'] == 'markdown':
        cell['source'] = [translate_to_korean(line) if translate_to_korean(line) is not None else line for line in cell['source']]

# Save the translated notebook
translated_name = target_language + '_' + notebook_name[:-6] + '.ipynb'
with open(translated_name, 'w', encoding='utf-8') as file:
    json.dump(notebook, file, ensure_ascii=False, indent=4)
```
